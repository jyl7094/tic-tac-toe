```python
def display_board(board):
    print(f' {board[6]} | {board[7]} | {board[8]} \n'
           '-----------\n'
          f' {board[3]} | {board[4]} | {board[5]} \n'
           '-----------\n'
          f' {board[0]} | {board[1]} | {board[2]} \n')
```

```python
def get_input(board, player, player_string):
    user_input = ''
    print(f'{player_string}\'s turn.')
    while not user_input:
        try:
            user_input = int(input('Enter a number from 1-9: '))
            print()
            if not user_input in range(1, 10):
                user_input = ''
                raise
            if board[user_input - 1] != ' ':
                user_input = ''
                raise
            return (player, user_input)
        except:
            print('Try again: Invalid input.')
```

```python
def update_board(board, player):
    if player[0] == 'p1':
        board[player[1] - 1] = 'O'
    else:
        board[player[1] - 1] = 'X'
    
    return board
```

```python
def check_condition(board, winner):
    p1_victory = ['O', 'O', 'O']
    p2_victory = ['X', 'X', 'X']
    if board[0:3] == p1_victory or \
       board[3:6] == p1_victory or \
       board[6:] == p1_victory or \
       board[::3] == p1_victory or \
       board[1::3] == p1_victory or \
       board[2::3] == p1_victory or \
       board[0::4] == p1_victory or \
       board[2:7:2] == p1_victory:
        winner = 'p1'
    elif board[0:3] == p2_victory or \
       board[3:6] == p2_victory or \
       board[6:] == p2_victory or \
       board[::3] == p2_victory or \
       board[1::3] == p2_victory or \
       board[2::3] == p2_victory or \
       board[0::4] == p2_victory or \
       board[2:7:2] == p2_victory:
        winner = 'p2'
    else:
        winner = ''
    
    return winner
        
```

```python
def setup():
    rounds = 1
    winner = ''
    players = ('p1', 'p2')
    players_string = {
        'p1': 'Player 1',
        'p2': 'Player 2'
    }
    player_tracker = 0
    board = [' '] * 9
    while not winner:
        if rounds == 10:
            print('\nIt\'s a draw!')
            break
        display_board(board)
        print('-' * 30)
        print(f'Round {rounds}'.center(30))
        print()
        p1ayer = get_input(board, players[player_tracker], players_string[players[player_tracker]])
        board = update_board(board, p1ayer)
        winner = check_condition(board, winner)
        if winner:
            print(f'\nThe winner is {players_string[winner]}!\n')
            display_board(board)
        player_tracker = 1 if player_tracker == 0 else 0
        rounds += 1
```

```python
def play():
    playing = True
    print('Demo of Tic-Tac-Toe game Made by Jay Lee (GitHub: jyl7094)')
    while playing:
        response = ''
        print()
        setup()
        while not response:
            response = input('To keep playing, enter \'Yes\'. Otherwise, enter \'No\' to exit: ')
            if response == 'Yes':
                break
            elif response == 'No':
                print('\nThanks for playing!')
                playing = False
                break
            else:
                response = ''
```

```python
play()
```
